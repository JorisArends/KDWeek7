# Dataset ophalen eredivisie


```python
import requests
import pandas as pd
def sparql_query(query):
    r = requests.get('https://query.wikidata.org/sparql', 
                 params = {'format': 'json', 'query': query})
    return r.json()
```

### Om het verband te onderzoeken tussen de wellicht negatieve invloed van een hoge mate van populariteit op social media kanalen op de prestaties van de speler in kwestie; en het mogelijke verband tussen de prestaties van spitsen en middenvelders op het veld (doelpunten) en hun populariteit op social media, maken wij gebruik van de data van alle Eredivisieclubs op WikiData. 

### Zoals je hieronder kunt zien hebben wij een query geschreven waarmee we de spelernaam, clubnaam, het aantal wedstrijden, goals en Twittervolgers, de positie op het veld en de Instagram- en Twitter gebruikersnaam ophalen van alle Eredivisiespelers. Deze data wordt weergegeven in een dataframe waarmee we vervolgens onze verdere analyses kunnen uitvoeren. 

# Data waarmee we gaan werken

```python
output = sparql_query("""
SELECT DISTINCT ?spelerLabel ?clubLabel ?matches ?goals ?nationalityLabel ?twitter_name ?twitter_followers ?instagram_name ?instagram_followers 
WHERE
{
    ?speler p:P54 ?statement.
    ?statement ps:P54 ?club.
    ?club wdt:P31 wd:Q476028 ;
        wdt:P118 wd:Q167541 .
        FILTER (NOT EXISTS {?club wdt:P576 ?X.})
    ?statement pq:P1350 ?matches .
    ?statement pq:P1351 ?goals .
        
    OPTIONAL{ ?speler wdt:P2003 ?instagram_name .}
    
    ?speler p:P2002 ?twitter .
    ?speler wdt:P27 ?nationality .
    ?twitter ps:P2002 ?twitter_name .
    ?twitter pq:P3744 ?twitter_followers .
     SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],nl,en". }
}""")

```

Voorbereiden van de Dataframe op basis van querydata

```python
result = [{x: record[x]['value'] for x in record} for record in output['results']['bindings']]
result_df = pd.DataFrame(columns=output['head']['vars'], data=result)
result_df = result_df.set_index(['spelerLabel']).sort_index()
df = result_df.copy()

# Format de dataframe op de juiste manier zodat duplicates worden samengevoegd. 
df[['matches', 'goals', 'twitter_followers', 'instagram_followers']] = df[['matches', 'goals', 'twitter_followers', 'instagram_followers']].apply(pd.to_numeric)

df = df.groupby(['spelerLabel']).agg({'clubLabel':', '.join, 'matches':'sum', 'goals':'sum', 
                                    'twitter_followers':'max', 'instagram_followers':'max'})

# Zorg ervoor dat de unieke kolommen terugkomen 
df[['nationalityLabel', 'instagram_name', 'twitter_name']] = result_df[~result_df.index.duplicated(keep='first')][['nationalityLabel', 'instagram_name', 'twitter_name']]

# Zet de string met clubs om tot een lijst en dan een set 
df['clubLabel'] = df['clubLabel'].str.split(', ')
df['clubLabel'] = df['clubLabel'].apply(lambda x : set(x))
df = df.reindex(result_df.columns, axis=1)

df.head(5)
```

# Correlatie social media en gemiddelde score

```python
df_math = df[['matches', 'goals', 'twitter_followers', 'instagram_followers']]
df_math = df_math.assign(mean_followers=df_math.loc[:, ["twitter_followers", "instagram_followers"]].mean(axis=1))
df_math['score'] =  (df_math['goals'] / df_math['matches'])

df_math.head(5)
```

```python
correlation_df = {'twitter_followers': df_math['twitter_followers'],
                  'instagram_followers': df_math['instagram_followers'],
                  'mean_followers': df_math['mean_followers'],
                  'score': df_math['score']}

cor_df = pd.DataFrame(correlation_df,columns=['twitter_followers','instagram_followers', 'mean_followers', 'score'])
cor_matrix = cor_df.corr()

display(cor_matrix)
```

```python
import plotly.express as px
fig = px.scatter(df_math, x="twitter_followers", y="score",
           size="twitter_followers", color="score", hover_name=df_math.index, log_x=True, color_continuous_scale=px.colors.sequential.Viridis
                )
fig.update_layout(
    title_text="Eredivisie spelers met hun score en twitter volgers"
)


fig.show()
```
